```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib
```

```python
buildings = pd.read_csv("https://github.com/UIUC-iSchool-DataViz/is445_data/raw/main/building_inventory.csv",
           na_values = {'Square Footage': 0,
                       'Year Acquired': 0,
                       'Year Constructed': 0}) 
```

```python
buildings.head()
```

```python
stats = buildings.groupby('Year Acquired')['Square Footage'].describe()
stats
```

```python
fig, ax = plt.subplots(figsize=(10,2))
ax.plot(stats['max'], marker='.', linewidth=1.0)
ax.set_xlabel('Year Acquired')
ax.set_ylabel('Square Footage')
ax.set_yscale('log')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,2))
ax.plot(stats['max'], marker='.', linewidth=1.0)
ax.plot(stats['min'], marker='.', linewidth=1.0)
ax.set_xlabel('Year Acquired')
ax.set_ylabel('Square Footage')
ax.set_yscale('log')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,2))
ax.plot(stats['max'], marker='.', linewidth=1.0, label='Max')
ax.plot(stats['min'], marker='.', linewidth=1.0, label='Min')
ax.set_xlabel('Year Acquired')
ax.set_ylabel('Square Footage')
ax.set_yscale('log')
ax.legend()
plt.show()
```

```python
plt.style.use('ggplot') # sets the plotting style for all plots
fig, ax = plt.subplots(figsize=(10,2))
ax.plot(stats['max'], marker='.', linewidth=1.0, label='Max')
ax.plot(stats['min'], marker='.', linewidth=1.0, label='Min')
ax.set_xlabel('Year Acquired')
ax.set_ylabel('Square Footage')
ax.set_yscale('log')
ax.legend()
plt.show()
```

```python
# we want two elements of interactivity
# 1. colormap
# 2. log (on/off) of color map
```

```python
# Step 1 -- colormap: make a plot with parameter hard-coded
colormap = 'Reds'
plt.imshow(michigan, cmap=colormap)
plt.colorbar()
plt.show()
```

```python
# Step 2 -- colormap: make a function with that parameter as an input
#colormap = 'Reds'
def make_mich_plot(colormap):
    plt.imshow(michigan, cmap=colormap)
    plt.colorbar()
    plt.show()
```
