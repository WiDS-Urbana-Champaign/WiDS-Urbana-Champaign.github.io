```python
import pandas as pd
```

```python
#buildings = pd.read_csv('https://github.com/UIUC-iSchool-DataViz/is445_data/raw/main/building_inventory.csv')
buildings = pd.read_csv('https://github.com/UIUC-iSchool-DataViz/is445_data/raw/main/building_inventory.csv',
                       na_values = {'Square Footage':0, # we probably want to look into this in detail!
                                    'Year Acquired':0,
                                    'Year Constructed':0
                       })
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
import ipywidgets
```

```python
stats = buildings.groupby('Year Acquired')['Square Footage'].describe()
stats
```

```python
#plt.style.use('ggplot')
@ipywidgets.interact(style=plt.style.available)
def make_plot(style):
    with plt.style.context(style):
        # CMD + ]
        fig, ax = plt.subplots(figsize=(10,2))
        ax.plot(stats['max'], marker='.', label='Max')
        ax.plot(stats['min'], marker='.', label='Min')
        ax.set_xlabel('Year (Acquired)')
        ax.set_ylabel('Square Footage')
        ax.set_yscale('log')
        ax.legend()
        plt.show()
```
