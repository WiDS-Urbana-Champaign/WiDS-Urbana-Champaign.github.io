````python
fig, ax = plt.subplots(figsize=(10,3))
ax.hist(im_data[:,:,0]) # [:,:,0] == all pixels in rows, and columns, only the Red color channel
plt.show()
````

````python
fig, ax = plt.subplots(figsize=(10,3))
ax.hist(im_data[:,:,0].flatten(), bins=256) # [:,:,0] == all pixels in rows, and columns, only the Red color channel
plt.show()
````

````python
fig, ax = plt.subplots(figsize=(10,3))
ax.hist(im_data[:,:,0].flatten(), bins=256) # [:,:,0] == all pixels in rows, and columns, only the Red color channel
ax.set_xlabel('Red channel bins')
ax.set_ylabel('# of pixels per bin')
plt.show()
````

````python
fig, ax = plt.subplots(figsize=(10,3))
ax.hist(im_data[:,:,0].flatten(), bins=256, color='red') # [:,:,0] == only the Red color channel
ax.hist(im_data[:,:,1].flatten(), bins=256, color='green') # [:,:,1] == only the Green color channel
ax.set_xlabel('Red channel bins')
ax.set_ylabel('# of pixels per bin')
plt.show()
````

````python
fig, ax = plt.subplots(figsize=(10,2))
ax.hist(im_data[:,:,0].flatten(), bins=256, color='red') # [:,:,0] == only the Red color channel
ax.hist(im_data[:,:,1].flatten(), bins=256, color='green') # [:,:,1] == only the Green color channel
ax.hist(im_data[:,:,2].flatten(), bins=256, color='blue') # [:,:,1] == only the Blue color channel
ax.hist(im_data[:,:,3].flatten(), bins=256, color='gray') # [:,:,1] == only the Alpha/opacity color channel
ax.set_xlabel('Channel bins')
ax.set_ylabel('# of pixels per bin')
plt.show()
````

````python
fig, ax = plt.subplots(4, 1, figsize=(10,8))
ax[0].hist(im_data[:,:,0].flatten(), bins=256, color='red') # [:,:,0] == only the Red color channel
ax[1].hist(im_data[:,:,1].flatten(), bins=256, color='green') # [:,:,1] == only the Green color channel
ax[2].hist(im_data[:,:,2].flatten(), bins=256, color='blue') # [:,:,1] == only the Blue color channel
ax[3].hist(im_data[:,:,3].flatten(), bins=256, color='gray') # [:,:,1] == only the Alpha/opacity color channel

# add some x/y labels
ax[0].set_xlabel('Red channel bins')
ax[1].set_xlabel('Green channel bins')
ax[2].set_xlabel('Blue channel bins')
ax[3].set_xlabel('Alpha channel bins')
# y-axis will be the same, do a for loop
for a in ax:
    a.set_ylabel('# of pixels')

fig.tight_layout() # fits things nicely in a small space w/o having overlaps
plt.show()
````

````python
fig, ax = plt.subplots(5, 1, figsize=(10,8)) # add something in for the image
ax[0].hist(im_data[:,:,0].flatten(), bins=256, color='red') # [:,:,0] == only the Red color channel
ax[1].hist(im_data[:,:,1].flatten(), bins=256, color='green') # [:,:,1] == only the Green color channel
ax[2].hist(im_data[:,:,2].flatten(), bins=256, color='blue') # [:,:,1] == only the Blue color channel
ax[3].hist(im_data[:,:,3].flatten(), bins=256, color='gray') # [:,:,1] == only the Alpha/opacity color channel

# add some x/y labels
ax[0].set_xlabel('Red channel bins')
ax[1].set_xlabel('Green channel bins')
ax[2].set_xlabel('Blue channel bins')
ax[3].set_xlabel('Alpha channel bins')
# y-axis will be the same, do a for loop
for a in ax[:-1]: # make sure no label for bottom plot
    a.set_ylabel('# of pixels')

ax[4].imshow(im_data)

fig.tight_layout() # fits things nicely in a small space w/o having overlaps
plt.show()
````
